```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
```

```python
df = pd.read_csv("cleaned_customer_churn.csv")
```

```python
churn_counts = df['Churn Label'].value_counts()
print(churn_counts)

churn_rate = df['Churn Label'].value_counts(normalize=True) * 100
print(churn_rate)
```

```python
#visualization
sns.countplot(x='Churn Label', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
df['Churn Flag'] = df['Churn Value']
```

```python
#Business Insight

#This means:

#Around 1 in 4 customers leave
###Churn is significant enough to impact revenue
#Retention strategy can create strong business value
```

```python
# Churn by Contract Type
```

```python
contract_churn = pd.crosstab(df['Contract'], df['Churn Label'], normalize='index') * 100
print(contract_churn)
```

```python
contract_churn.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("Churn by Contract Type")
plt.ylabel("% Customers")
plt.show()
```

```python
#3. Churn by Tenure Group
```

```python
tenure_churn = pd.crosstab(df['TenureGroup'], df['Churn Label'], normalize='index') * 100
print(tenure_churn)
```

```python
tenure_churn.plot(kind='bar', stacked=True, figsize=(8,5))
plt.title("Churn by Tenure Group")
plt.show()
```

```python
#insight: New customers (0-12 months) have the highest churn rate, indicating early dissatisfaction or unmet expectations.
```

```python
#4. Monthly Charges vs Churn
```

```python
sns.boxplot(x='Churn Label', y='Monthly Charges', data=df)
plt.title("Monthly Charges by Churn")
plt.show()
```

```python
#insight:Do churned customers pay more? Yes, they have higher monthly charges on average, which may contribute to their decision to leave.
```

```python
#5. Total Charges vs Churn
```

```python
sns.boxplot(x='Churn Label', y='Total Charges', data=df)
plt.title("Total Charges by Churn")
plt.show()
```

```python
#10. Correlation Heatmap
```

```python
num_cols = ['Tenure Months', 'Monthly Charges', 'Total Charges', 'Churn Flag', 'RiskScore']
corr = df[num_cols].corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
#insights:
# Long-tenure customers are less likely to churn.
#New customers show higher churn risk.
#RiskScore has meaningful alignment with churn (0.51 correlation).
#Revenue accumulates strongly with customer retention.
#Monthly charges alone are a weaker churn driver than tenure/contract behavior.
```
